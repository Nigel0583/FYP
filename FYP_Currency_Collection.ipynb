```python
import numpy as np
import pandas as pd
import pandas_datareader as web
import plotly.graph_objs as go
#Data Source
import yfinance as yf
import datetime
```

```python
data = yf.download(tickers = 'GBPEUR=X' ,start="2019-01-01", end="2020-12-31", interval = '1d')
start = datetime.datetime(2019, 1, 1)
end = datetime.datetime(2020, 12, 31)
df = web.DataReader("GBPEUR=X",'yahoo', start, end)

#df.to_csv('yahoo_GBPEUR.csv')
```

```python
#declare figure
fig = go.Figure()

#Candlestick
fig.add_trace(go.Candlestick(x=data.index,
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close'], name = 'market data'))

# Add titles
fig.update_layout(
    title='GBP/EUR')

#Show
fig.show()
```

```python
GBPEUR_data = pd.read_csv("Brexit_Tweets_19_20/yahoo_GBPEUR.csv", parse_dates=["Date"])
tweets_data = pd.read_csv("Brexit_Tweets_19_20/tweets_sentiments.csv", parse_dates=["Created At"])
```

```python
tweets_day = tweets_data.set_index(pd.to_datetime(tweets_data['Created At'])).resample('D').mean()
tweets_day['Date'] = tweets_day.index
currency_sentiments = GBPEUR_data.merge(tweets_day)
currency_sentiments.Polarity.fillna(currency_sentiments.Polarity.mean(), inplace=True)
```
